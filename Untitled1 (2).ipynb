We'll use Levene's test to examine the potential interplay between months, booking changes, and cancellations. The reason we are comparing booking changes is to try and see seasonality/season volatility, when many people are hecticly planning and things change. This can introduce volatility and irregularties in our dataset, that if somehwat ordered, we may need to structure our models around. This analysis aims to determine if these factors are actually connected. While prices might independently influence cancellations, we're interested in discerning if seasonality or holiday volatility plays a role. If Levene's test reveals significant associations among months, booking changes, and cancellations, it suggests a unique insight into how hotel managers can strategize for specific time periods.

```python
stat_adr, p_adr = levene(*[group['adr'] for name, group in df.groupby('arrival_date_month')])
print('Levene’s test for adr (Statistic: {:.3f}, P-value: {:.3f})'.format(stat_adr, p_adr))

# Levene's test for equality of variances in booking_changes
stat_bc, p_bc = levene(*[group['booking_changes'] for name, group in df.groupby('arrival_date_month')])
print('Levene’s test for booking changes (Statistic: {:.3f}, P-value: {:.3f})'.format(stat_bc, p_bc))

# Decision based on Levene's test
if p_adr > 0.05 and p_bc > 0.05:
    print("Variances are equal across groups. Proceed with ANOVA.")
else:
    print("Variances are not equal. Consider using transformations or non-parametric tests.")
```

Looks like thay have statiscally notable unequal variance levels, suggesting that their variances might be less prone to comparison. Therefore, it's likely theyour models should account for different levels of variance.

Why do a Chi-Square test?

A chi-square test is the test of choice for seeing whether one variable relates to changes in another variable, specfically it can help idenfity if some changes are due to change or if they are due to just happenstance.

As a quick intutiive way to udnerstand it, it measure the difference between if we observed the frequencies of data assuming the variables were related, contrasted to if we observe the data assuming there is NO correlation. Here is a short youtube video that gets the point across:

https://www.youtube.com/watch?v=7_cs1YlZoug


```python
contingency_table = pd.crosstab(df['country'], df['is_canceled'])

# Perform the chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p_value}")
```

Let's look at the numbers: that number in speicifc is huge: it reprsents teh difference between no-correlation ie assuming correlation: meaning for our suggestion that country andd cancellations are related, this is a great indicator. But how do we know whether this is by chance or not?

As we covered before in this tutorial, the P value once again shows that this is not really a chance or coincedence, that this relation is real and is actually pretty strong. We can also do a Z test like we did last time to confirm this.


Seems like the answer is OVERWHELIMINGLY YES. Now, let's check out if Portugal in specific is an indicator of a possible cancellation.

```python
# Counts of cancellations
count_prt = prt_data['is_canceled'].sum()
count_non_prt = non_prt_data['is_canceled'].sum()

# Number of observations
nobs_prt = prt_data['is_canceled'].count()
nobs_non_prt = non_prt_data['is_canceled'].count()

# Z-test for two proportions
stat, pval = proportions_ztest([count_prt, count_non_prt], [nobs_prt, nobs_non_prt])

# Print the results
print(f"Z-statistic: {stat}")
print(f"P-value: {pval}")
```

Let's start with Logistic Regression. What is it? To keep it short, it works the way any of our equations in algebra do: we have a Dependent variable and then a certain amount of predictor variables that go into an equation to determine whether something can be classified one way or not. The way that this model trains, is that for every data point it makes a guess, and it tries to optimize its guess to be correct. It adjusts it's weights for predictor variables to try and be correct as much as possible. As such it's prone to overfitting since its literaly jsut adjusting to the numbers in a dataset. However, it's a good start.

I'll only repeat this once because the SKLearn Machine Learning models are prettty simple to use. Because we have done the dirty work of setting up and partitioning the data, all we have to dois just call.fit to fir the data/train it, and then .predict to predict for our entire test dataset. Then, we can compare our predictions to the actual values for the various metrics we want.


```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)

y_pred_lr = logistic_model.predict(X_test)

acc_lr = accuracy_score(y_test, y_pred_lr)
conf = confusion_matrix(y_test, y_pred_lr)
clf_report = classification_report(y_test, y_pred_lr)

print(f"Accuracy Score of Logistic Regression is : {acc_lr}")
print(f"Confusion Matrix : \n{conf}")
print(f"Classification Report : \n{clf_report}")
```
